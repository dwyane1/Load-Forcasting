Implementation of Load Forecasting(Day) using **Moving Average** on **Day Data**

```
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import io
```

```
from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_csv(io.BytesIO(uploaded['D_load.csv']))
plt.figure(figsize=(20,5))
plt.plot(df[:500], label='Load')
plt.legend(loc='upper left')
plt.xlabel("Hours")
plt.ylabel("Load")
plt.show()
```

```
def mv_avg (l,n=-1): 
    sum=0
    if n<=-0:
         return 0
    
    for i in range(n):
        sum+=l[i]
    return round(sum/n)

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```
l=list(df['Load']) #converting into a list


l_mv_avg=[]
for n in range(len(l)):
  l_mv_avg.append(mv_avg(l,n))

```

```
plt.figure(figsize=(10,6))
plt.plot(l[0:100], label='Load')
plt.plot(l_mv_avg[0:100], label='Moving Average', color='Red')
plt.legend(loc='upper left')
plt.xlabel("Hours")
plt.ylabel("Load")
plt.show()
```

```

mape=mean_absolute_percentage_error(l, l_mv_avg)
print("Mean Absolute Error Percentage in Moving Average Method is ", round(mape,2),"%")

```
